```python
from openai import OpenAI
import os
```

```python
client = OpenAI(
    api_key=os.environ.get("OPENAI_API_KEY"))

#API confing
with open("texts_and_prompts.json", "r") as file:
    data = json.load(file)
    
texts = data["texts"]
prompts = data["prompts"]

#Models to tests
models = ["gpt-4o", "gpt-4o-mini", "o1-mini", "gpt-4-turbo"]

#Initialazing list to store results
results = []

#Combinations loop
for model in models:
    for prompt in prompts:
        for text in texts:
            try:
                #Model input
                message = prompt + text

                #OpenAI API calling
                if model == "o1-mini":
                    response = client.chat.completions.create(
                        model=model,
                        messages=[{"role": "user", "content": message}]
                        )
                else:
                    response = client.chat.completions.create(
                        model=model,
                        messages=[{"role": "user", "content": message}],
                        temperature=0.5
                        )

                # storing results
                results.append({
                    "model": model,
                    "prompt": prompt,
                    "text": text,
                    "output": response.choices[0].message.content
                })

            except Exception as e:
                #Ignoring errors
                results.append({
                    "model": model,
                    "prompt": prompt,
                    "text": text,
                    "error": str(e)
                })

print("Iteration completed successfully")
```

```python
# Saving results in JSON file
with open("Testing_results.json", "w") as file:
    json.dump(results, file, indent=4)

#Dataframing results
df = pd.DataFrame(results)
df.to_csv("Testing_results.csv", index=False)

print("Testing completed. Results saving in 'Testing_results.json' and 'Testing_results.csv'.")
```
